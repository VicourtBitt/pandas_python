```python
import pandas as pd
```

```python
csv_path = ".//pokemon.csv"

df = pd.read_csv(csv_path, index_col="#")
df
```

```python
greatAttack = df.loc[df['Attack'] > 150]
greatAttack.count()
```

```python
lowestSpeed = df.loc[df['Speed'] <= 10]
lowestSpeed.count()
```

```python
lowestSDef = df.loc[df['Sp. Def'] <= 25]
# lowestSDef = df.query("`Sp. Def` <= 25")
# lowestSDef = (df['Sp. Def'] <= 25).sum()

lowestSDef.count()
```

```python
legendaries = df.loc[df['Legendary'] == True]
# legendaries = df.query('Legendary == True')
legendaries
```

```python
fireFlying = df.loc[
    (df['Type 1'] == 'Fire') &
    (df['Type 2'] == 'Flying')
]

fireFlying
```

```python
poisonous = df.loc[
    (df['Type 1'] == 'Poison') |
    (df['Type 2'] == 'Poison')
]

poisonous
```

```python
iceWall = df.loc[
    df['Type 1'] == 'Ice'
]

iceWall.sort_values(by='Defense', ascending=False).iloc[0, 0]
```
